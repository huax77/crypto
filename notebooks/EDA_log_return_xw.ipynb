```python
import warnings
warnings.filterwarnings('ignore')

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.stattools import adfuller, kpss

# -----------------------------------------------------------------------------
# Configuration
# -----------------------------------------------------------------------------
# Notebook is in EDA/, data/ is its sibling
DATA_ROOT   = os.path.join("..", "data", "resampled")
SPLIT_LABEL = "2025-01-01"   # fixed-date split

INTERVALS = ["1min", "10min", "1h", "1d"]
FREQ_MAP  = {"1min": "T", "10min": "10T", "1h": "H", "1d": "D"}

WINDOW_DAYS = 90   # keep last 90 days of returns
MAX_LAGS    = 20   # for ACF/PACF

# -----------------------------------------------------------------------------
# Helper functions
# -----------------------------------------------------------------------------
def load_train(interval: str) -> pd.DataFrame:
    """Load train_{interval}.parquet and ensure timestamp dtype."""
    path = os.path.join(DATA_ROOT, f"train_{interval}.parquet")
    df = pd.read_parquet(path)
    df["timestamp"] = pd.to_datetime(df["timestamp"])
    return df

def preprocess_log_return(df: pd.DataFrame, interval: str) -> pd.Series:
    """
    - Reindex to strict freq
    - Compute log returns
    - Truncate to the last WINDOW_DAYS
    """
    series = (
        df
        .set_index("timestamp")["close"]
        .asfreq(FREQ_MAP[interval])
        .dropna()
    )
    lr = np.log(series).diff().dropna()
    cutoff = lr.index.max() - pd.Timedelta(days=WINDOW_DAYS)
    return lr[lr.index >= cutoff]

def descriptive_stats(s: pd.Series) -> dict:
    """Basic stats for a Series."""
    return {
        "mean":     s.mean(),
        "std":      s.std(),
        "skew":     s.skew(),
        "kurtosis": s.kurtosis(),
        "min":      s.min(),
        "25%":      s.quantile(0.25),
        "50%":      s.median(),
        "75%":      s.quantile(0.75),
        "max":      s.max(),
    }

def stationarity_tests(s: pd.Series) -> dict:
    """Return ADF and KPSS p-values."""
    return {
        "adf_p":  adfuller(s, autolag="AIC")[1],
        "kpss_p": kpss(s, nlags="auto")[1],
    }

def eda_task(interval: str, coin: str):
    """
    Perform EDA for one coin + interval:
      - print stats & p-values
      - show histogram, ACF, PACF inline
    """
    df      = load_train(interval)
    df_coin = df[df["coin_id"] == coin]
    lr      = preprocess_log_return(df_coin, interval)

    stats = descriptive_stats(lr)
    stats.update(stationarity_tests(lr))
    print(f"\n=== EDA: {SPLIT_LABEL} | {interval} | {coin} ===")
    for k, v in stats.items():
        print(f"{k}: {v:.4f}" if isinstance(v, float) else f"{k}: {v}")

    plt.figure(figsize=(6,4))
    lr.hist(bins=50)
    plt.title(f"{coin} @ {interval}  Log-return Histogram")
    plt.show()

    plt.figure(figsize=(6,4))
    plot_acf(lr, lags=MAX_LAGS, title=f"{coin} @ {interval}  ACF")
    plt.show()

    plt.figure(figsize=(6,4))
    plot_pacf(lr, lags=MAX_LAGS, method="ywm", title=f"{coin} @ {interval}  PACF")
    plt.show()

# -----------------------------------------------------------------------------
# Main EDA loop
# -----------------------------------------------------------------------------
if __name__ == "__main__":
    for interval in INTERVALS:
        df = load_train(interval)
        for coin in sorted(df["coin_id"].unique()):
            eda_task(interval, coin)
```
